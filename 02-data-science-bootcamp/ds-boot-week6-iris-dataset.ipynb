we're going to use a package called plotnine to produce a 2 dimensional graph

```python
from plotnine import *

ggplot(iris, aes(x='sepallength', y='sepalwidth')) + geom_point()
```

let's try petal length/petal width 

```python
ggplot(iris, aes(x='petallength', y='petalwidth')) + geom_point()
```

shows a clear relationship between two values, however, not clear which observations belong to each species of flower

```python
ggplot(data = iris) +geom_point(mapping = aes(x='sepallength', y='sepalwidth', color='class'))
```

```python
ggplot(data = iris) +geom_point(mapping = aes(x='petallength', y='petalwidth', color='class'))
```
